```python
import numpy as np
import sklearn
from sklearn.metrics import mean_squared_error
from tqdm import tqdm_notebook
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
class Univeratelinearregression:
  def __init__(self): # instrector class __init__
    self.theta_0 = None # theta_0 (the intercept) and theta_1 (the slope). , initialy nune value
    self.theta_1 = None # None because the model is not yet trained.

  def hypothesis (self,x):
    return self.theta_0 + self.theta_1 * x # h(x)

  def  gradient_theta_0 (self , x , y): # equation above img
    ypred = self.hypothesis (x)
    return (ypred - y)

  def gradient_theta_1 (self , x , y):
    ypred = self.hypothesis (x)
    return (ypred - y) * x # theta 1 * x in the equation

  def fit (self,x,y,epochs = 1 ,learning_rate = 0.01):

    self.theta_0 = 0.0 # initialise 0
    self.theta_1 = 0.0
    m = x.shape[0] # number of training set
    for i in tqdm_notebook (range (epochs),total = epochs , unit = "epoch"): #motham epoches
      dtheta_0 = 0.0
      dtheta_1 = 0.0

      for X,Y in zip (x,y):
        dtheta_0 = dtheta_0 + self.gradient_theta_0 (X,Y)
        dtheta_1 = dtheta_1 + self.gradient_theta_1 (X,Y)
      self.theta_0 = self.theta_0 - (learning_rate * (dtheta_0 / m))
      self.theta_1 = self.theta_1 - (learning_rate * (dtheta_1 / m))

  def predict (self,x):
    y_pred = []
    for x in x:
      y_pred.append(self.hypothesis(x))
    return np.array(y_pred)






```
