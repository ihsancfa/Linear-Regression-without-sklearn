```python
import numpy as np
from sklearn.metrics import mean_squared_error
from tqdm import tqdm_notebook
import matplotlib.pyplot as plt
```

```python
class UniveriateLinearRegression:

  def __init__(self):
    self.theta_0 = None
    self.theta_1 = None

  def hypothesis(self,x):
    return self.theta_0 + self.theta_1*x

  def grad_theta_0 (self,x,y):
    ypred = self.hypothesis(x)
    return (ypred-y)

  def grad_theta_1 (self,x,y):
    ypred = self.hypothesis(x)
    return (ypred-y)*x

  def fit(self,X,Y,epochs=1,learning_rate=.01):

    self.theta_0 = 0.0
    self.theta_1 = 0.0
    m = X.shape[0]
    for i in range(epochs):
      dtheta_0 = 0.0
      dtheta_1 = 0.0
      for x,y in zip(X,Y):
        dtheta_0 = dtheta_0 + self.grad_theta_0(x,y)
        dtheta_1 = dtheta_1 + self.grad_theta_1(x,y)
      self.theta_0 = self.theta_0 - learning_rate * dtheta_0 / m
      self.theta_1 = self.theta_1 - learning_rate * dtheta_1 / m

  def predict(self,X):
    y_pred = []
    for x in X:
      y_pred.append(self.hypothesis(x))
    return np.array(y_pred)
```

```python
data = np.loadtxt(open(r"T:\My Drive\000 ihsan\02 Courses\003GIT\MYProject\Linear_regression_without_sklearn\Data\salary_exp_amount.txt","rb"),delimiter = ",")
X_train = data[:,0]
Y_train = data[:,1]
X_train = X_train.reshape(-1,1)
Y_train = Y_train.reshape(-1,1)
```

```python
model = UniveriateLinearRegression()
```

```python
model.fit(X_train,Y_train,epochs=4000,learning_rate=.001)
```

```python
data_test = np.loadtxt(open(r"T:\My Drive\000 ihsan\02 Courses\003GIT\MYProject\Linear_regression_without_sklearn\Data\test salary data.txt","rb"),delimiter = ",")
X_test = data_test[:,0]
Y_test = data_test[:,1]
X_test = X_test.reshape(-1,1)
Y_test = Y_test.reshape(-1,1)
```

```python
Y_pred = model.predict(X_test)
```

```python
error = mean_squared_error(Y_pred,Y_test)
print(error)
```
